# Dizeleri Dilimleme (Slicing Strings)

Dilimleme syntax'ını kullanarak dizileri dilimleyebilir ve istediğimiz karakterlerini döndürebiliriz.

X[a,b]  = a indexinden b indeksine kadarki karakterleri al.


image.png

```python
b = "hello world"
print(b[2:4])
```

Başlangıç dizini girilmezse ilk karakterden bölmeye başlanır.

```python
b = "hello world"
print(b[:6])
```

Bitiş dizini verilmezse son karaktere kadar bölümlenir.

```python
b  = "hello world"
print(b[6:])
```

Hem başlangıç hem bitiş dizini verilmezse baştan sona diziyi döndürür.

```python
b = "hello world"
print(b[:])
```

Dizenin sonundan başlamak için negatif dilimleme kullanılır.

```python
b="hello world"
print(b[-5:-2])
```

Dilimleme atlama yaparak da gerçekleştirilebilir.

```python
b = "hello world"
print(b[0:12:2])
```

# Dizeleri Biçimlendirme (Strings Format)

Dizeler ve sayıları + ile birleştiremeyiz. Hata alırız.


```python
dize="banana"
sayı=4
print(sayı+dize)
```

format() kullanırsak dize ve sayıları birleştirebiliriz. {} işaretlerine yer tutucu denir. Yerine gelecek elemanın konumunu işaret eder. format() metodu bağımsız değişkenleri alır biçimlendirir ve yer tutucularının bulunduğu konuma yerleştirir.

```python
age=36
text = "i am {} years old"
print(text.format(age))
```

```python
age1 = 20
age2 = 40
text = "i am {} , and my father is {} years old. "
print(text.format(age1,age2))
```

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
```

Bağımsız değişkenlerin doğru yer tutuculara yerleştirildiğinden emin olmak için yer tutuculara numara verebilirsiniz.

```python
age1 = 20
age2 = 40
text = "i am {0} , and my father is {1} years old. My cousin is {0} years old "
print(text.format(age1,age2,age1))
```

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

format() içine parametreler key-value şeklinde de yerleştirilebilir.

```python
text = "i am {age1} , and my father is {age2} years old. My cousin is {age1} years old "
print(text.format(age1=20,age2=40))
```

# Kaçış Karakterleri (Escape Charachters)

Kaçış karakteri eklemek istediğiniz karakterin ardından gelen ters eğik çizgidir.

Çift tırnak içine alınmış bir dize içinde çift tırnak kullanırsanız bir hata alırsınız.

Ancak kaçış karakteri kullanırsanız hata almazsınız.

```python
txt = "We are the so-called "Vikings" from the north."
print(txt)
```

```python
txt = "We are the so-called \"Vikings\" from the north."
print(txt)
```

```python
txt = "We are the so-called \'Vikings\' from the north."
print(txt)
```

```python
txt = "We are the so-called \\Vikings\\ from the north."
print(txt)
```

```python
txt = "We are the so-called \nVikings\n from the north."
print(txt)
```

```python
txt = "We are the so-called \rVikings\r from the north."
print(txt)
```

```python
txt = "We are the so-called \tVikings\t from the north."
print(txt)
```

```python
txt = "We are the so-called \bVikings\b from the north."
print(txt)
```

```python
txt = "We are the so-called \fVikings\f from the north."
print(txt)
```

#### center(), doldurma karakteri olarak belirtilen bir karakteri (varsayılan olarak boşluktur) kullanarak dizeyi ortalayarak hizalar.  




**Sözdizimi**

string.center(length, character)

```python
txt = "banana"

x = txt.center(20, "O")

print(x)
```

#### count()dizede belirtilen bir değerin kaç kez göründüğünü döndürür.

**Sözdizimi**

string.count(value, start, end)

```python
txt = "I love apples, apple are my favorite fruit"

x = txt.count("apple", 10, 24)

print(x)
```

#### encode(), belirtilen kodlamayı kullanarak dizeyi kodlar. Kodlama belirtilmezse UTF-8 kullanılacaktır.

**Sözdizimi**

string.encode(encoding=encoding, errors=errors)

```python
txt = "My name is Ståle"

x = txt.encode()

print(x)
```

```python
#Bu örneklerde ascii kodlaması ve kodlanamayan bir karakter kullanılmış olup, sonuç farklı hatalarla gösterilmiştir
txt = "My name is Ståle"

print(txt.encode(encoding="ascii",errors="backslashreplace"))
print(txt.encode(encoding="ascii",errors="ignore"))
print(txt.encode(encoding="ascii",errors="namereplace"))
print(txt.encode(encoding="ascii",errors="replace"))
print(txt.encode(encoding="ascii",errors="xmlcharrefreplace"))
```
