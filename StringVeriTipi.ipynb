format() kullanrsak dize ve saylar birletirebiliriz. {} iaretlerine yer tutucu denir. Yerine gelecek elemann konumunu iaret eder. format() metodu bamsz deikenleri alr biimlendirir ve yer tutucularnn bulunduu konuma yerletirir.

```python
age=36
text = "i am {} years old"
print(text.format(age))
```

```python
age1 = 20
age2 = 40
text = "i am {} , and my father is {} years old. "
print(text.format(age1,age2))
```

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
```

Bamsz deikenlerin doru yer tutuculara yerletirildiinden emin olmak iin yer tutuculara numara verebilirsiniz.

```python
age1 = 20
age2 = 40
text = "i am {0} , and my father is {1} years old. My cousin is {0} years old "
print(text.format(age1,age2,age1))
```

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

format() iine parametreler key-value eklinde de yerletirilebilir.

```python
text = "i am {age1} , and my father is {age2} years old. My cousin is {age1} years old "
print(text.format(age1=20,age2=40))
```
