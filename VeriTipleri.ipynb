# VERİ TİPLERİ

Python’da, o anda elinizde bulunan bir verinin hangi tipte olduğunu bilmek son derece önemlidir. Çünkü bir verinin ait olduğu tip, o veriyle neler yapıp neler yapamayacağınızı belirler.

Programlamada veri türü önemli bir kavramdır.

Değişkenler farklı türdeki verileri depolayabilir ve farklı türler farklı şeyler yapabilir.

Python, bu kategorilerde varsayılan olarak aşağıdaki veri türlerini yerleşik olarak içerir

### Veri Tipini Öğrenme
Herhangi bir nesnenin veri türünü type() metodunu kullanarak alabilirsiniz

```python

var1 = "hello"
print(type(var1))

var2 = 10
print(type(var2))

var3=10.4
print(type(var3))
```

# Veri Tipini Ayarlama
Python'da, bir değişkene değer atadığınızda veri türü ayarlanır

```python
x = "hello world"
print(x)
print(type(x))
```

```python

x=20
print(x)
print(type(x))
```

```python
x=20.5
print(x)
print(type(x))
```

```python
x=1j
print(x)
print(type(x))
```

```python
x = ["apple", "banana", "cherry"]
print(x)
print(type(x))
```

```python
x=("apple","banana","cherry")
print(x)
print(type(x))
```

```python
x=range(6)
print(x)
print(type(x))
```

```python
x = {"name" : "John", "age" : 36}
print(x)
print(type(x))
```

```python
x= frozenset({"apple","banana","cherry"})
print(x)
print(type(x))
```

```python
x=True
print(x)
print(type(x))
```

```python
x=b"hello"
print(x)
print(type(x))
```

```python
x=bytearray(5)
print(x)
print(type(x))
```

```python
x=None
print(x)
print(type(x))
```

# Belirli Bir Veri Tipini Ayarlama
Veri türünü belirtmek istiyorsanız aşağıdaki yapıcı işlevlerini kullanabilirsiniz

```python
x = str("Hello World")
print(x)
print(type(x))
```

```python
x = int(20)
print(x)
print(type(x))
```

```python
x = float(20.5)
print(x)
print(type(x))
```

```python
x=complex(1j)
print(x)
print(type(x))
```

```python
x=list(("apple","banana","cherry"))
print(x)
print(type(x))
```

```python
x=tuple(("apple","banana","cherry"))
print(x)
print(type(x))
```

```python
x=range(6)
print(x)
print(type(x))
```

```python
x=dict(Name="John",Age=36)
print(x)
print(type(x))
```

```python
x=frozenset(("apple","banana","cherry"))
print(x)
print(type(x))
```

```python
#bool 0 HARİÇ İÇİNE YAZILAN HER DEĞERİ 1 OLARAK KABUL EDER.
x=bool(5)
print(x)
print(type(x))
```

```python
x=bytes(5)
print(x)
print(type(x))
```
