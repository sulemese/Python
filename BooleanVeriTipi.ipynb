Programlamada sıklıkla bir ifadenin doğru yada yanlış olması ile ilgileniriz.

Bool veri tipi de bu durumları tutar.

2 değere sahiptir. True yada false.

```python
a=True
b=False
```

True/False değerleri tırnak içinde yazılırsa bool değil string olarak değerlendirilir.

```python
a='True'
type(a)
```

```python
print(10>9)
print(10==9)
print(10<9)
print(not 10==9)
print(10 != 9)
print(not 10>9 )
```

if else gibi şartlı ifadelerin içine boolean sonuç döndüren mantıksal ifadeler koyulabilir.

```python
a=100
b=200

if b>a:
  print("B İS GREATER THEN A")
else:
  print("A İS GREATER THEN B")
```

Bool herhangi bir durumu değerlendirmemize ve karşılığında bize true/false dönmesine olanak tanır.

```python
print(bool("hello"))
print(bool(15))
```

```python
x="hello"
y=15
print(bool(x))
print(bool(y))
```

Boş string hariç tüm stringlerin bool değeri true'dur.


```python
print(bool("abc"))
print(bool(""))
```

Sıfır hariç tüm sayıların bool değeri true'dur.

image.png

```python
print(bool(10))
print(bool(0))
```

Boş olmayan tüm veri tiplerinin bool değeri true'dur. (list,tuple,dictionary,set vb.)

```python
print(bool(["apple", "cherry", "banana"]))
print(bool(()))
print(bool([]))
print(bool({}))
```

bool(False) ve bool(None) false'tur.

```python
bool(False)
bool(None)
```

Fonksiyonlar da bool değer döndürebilir.

```python
def myfunc():
  return True

print(myfunc())
```

```python
def myfunc():
  return True


if myfunc():
  print("YES")
else:
  print("NO")
```

Pythonda parametre olarak verilen nesnenin , parametre olarak verilen veri tipinde olup olmadığını kontrol etmek için "isinstance" fonksiyonu kullanıyoruz.

```python
x=200
print(isinstance(x,int))
print(isinstance(x,float))
print(isinstance(x,str))
```
