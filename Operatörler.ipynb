## 5. Kimlik Operatörleri
2 nesnenin aynı olup olmadığını ve aynı bellek konumuna sahip olup olmadığını karşılaştırır.

image.png

```python
#x ve y aynı değerlere sahip olduğundan x == y true döner.

x=["banana"]
y=["banana"]

print(x==y)
```

```python
#x ve y aynı değerlere sahip olmadığından x == y false döner.

x=["banana"]
y=["apple"]

print(x==y)
```

```python
#x ve y aynı konuma sahip olmadığından yani aynı nesne olmadığından x is y false döner.

x=["banana"]
y=["banana"]

print(x is y)
```

```python
#x ve y aynı konuma sahip ve aynı nesne olduğundan x is y true döner.

x=["banana"]
y=["banana"]

x = y

print(x is y)
```
