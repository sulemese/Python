# Tuple Güncelleme

Demetler değiştirilemez; yani demet oluşturulduktan sonra öğeleri değiştiremez, ekleyemez veya kaldıramazsınız.

Ancak bazı geçici çözümler var.

####1. Öğe Değiştirme

Bir demet oluşturulduktan sonra değerlerini değiştiremezsiniz. Tuple'lar değiştirilemez veya aynı zamanda denildiği gibi değişmezdir .

Ancak bir geçici çözüm var. Tuple'ı bir listeye dönüştürebilir, listeyi değiştirebilir ve listeyi tekrar Tuple'a dönüştürebilirsiniz.

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

####2. Öğe Ekleme

1. Yöntem : Tuple'ı Listeye Dönüştürmek

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
```

2. Yöntem : Tuple'a Tuple Eklemek

```python
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y

print(thistuple)
```

####3. Öğe Kaldırma

1. Yöntem : Tuple'ı bir listeye dönüştürürüz, öğeyi kaldırırız ve tekrar tuple'a dönüştürürüz.

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
```

2. Yöntem : Tuple'ı tamamen silebilirsiniz. Bunun için "del" anahtar kelimesini kullanabilirsiniz.

```python
thistuple = ("apple", "banana", "cherry")
del thistuple
```

# Tuple'ı Paketten Çıkartma (Unpack Tuples)

Bir tuple oluşturduğumuzda normalde ona değerler atarız. Buna bir demetin "paketlenmesi" denir.


```python
#Bir tuple'ı paketleme
fruits = ("apple", "banana", "cherry")
```

Ancak Python'da değerleri tekrar değişkenlere çıkarmamıza da izin veriliyor. Buna "paketten çıkarma" denir.

```python
fruits = ("apple", "banana", "cherry")

a,b,c = fruits
print(a,b,c)
```

Değişken sayısı değer sayısından azsa değişken adına bir yıldız (*) ekleyebilirsiniz; değerler değişkene bir liste halinde atanacaktır.

```python
fruits = ("apple", "banana", "cherry")

a,b,*c = fruits

print(c)
```

Yıldız işareti son değişkenden başka bir değişken adına eklenirse, Python, kalan değer sayısı kalan değişken sayısıyla eşleşene kadar değişkene değerler atayacaktır.

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

a, *b, c = fruits

print(a)
print(b)
print(c)
```

# Tuple Döngüsü

Bir for döngüsü kullanarak tuple öğeleri arasında geçiş yapabilirsiniz

```python
tuple1=(2,1,5,7,4)
for x in tuple1:
  print(x)
```

Ayrıca dizin numaralarına başvurarak tuple öğeleri arasında geçiş yapabilirsiniz.

Uygun bir dizin oluşturmak için range()ve len() işlevlerini kullanın.

```python
tuple1=("a","b","c")
for x in range(len(tuple1)):
  print(tuple1[x])
```

Bir while döngüsü kullanarak tuple öğeleri arasında geçiş yapabilirsiniz

Tuple'ın uzunluğunu belirlemek için len() işlevi kullanın  ardından 0'dan başlayın ve dizinlerine başvurarak tuple öğeleri arasında döngü yapın.

Her yinelemeden sonra dizini 1 artırmayı unutmayın.

```python
i=0
tuple1=("a","b","c")

while i<len(tuple1):
  print(tuple1[i])
  i=i+1
```

#Join Tuples

İki veya daha fazla tuple'ı birleştirmek için + operatörü kullanabilirsiniz.

```python
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

Bir demetin içeriğini belirli sayıda çarpmak istiyorsanız * operatörü kullanabilirsiniz.

```python
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```
