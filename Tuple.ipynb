# Tuple' Paketten kartma (Unpack Tuples)

Bir tuple oluturduumuzda normalde ona deerler atarz. Buna bir demetin "paketlenmesi" denir.


```python
#Bir tuple' paketleme
fruits = ("apple", "banana", "cherry")
```

Ancak Python'da deerleri tekrar deikenlere karmamza da izin veriliyor. Buna "paketten karma" denir.

```python
fruits = ("apple", "banana", "cherry")

a,b,c = fruits
print(a,b,c)
```

Deiken says deer saysndan azsa deiken adna bir yldz (*) ekleyebilirsiniz; deerler deikene bir liste halinde atanacaktr.

```python
fruits = ("apple", "banana", "cherry")

a,b,*c = fruits

print(c)
```

Yldz iareti son deikenden baka bir deiken adna eklenirse, Python, kalan deer says kalan deiken saysyla eleene kadar deikene deerler atayacaktr.

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

a, *b, c = fruits

print(a)
print(b)
print(c)
```

# Tuple Dngs

Bir for dngs kullanarak tuple eleri arasnda gei yapabilirsiniz

```python
tuple1=(2,1,5,7,4)
for x in tuple1:
  print(x)
```

Ayrca dizin numaralarna bavurarak tuple eleri arasnda gei yapabilirsiniz.

Uygun bir dizin oluturmak iin range()ve len() ilevlerini kullann.

```python
tuple1=("a","b","c")
for x in range(len(tuple1)):
  print(tuple1[x])
```

Bir while dngs kullanarak tuple eleri arasnda gei yapabilirsiniz

Tuple'n uzunluunu belirlemek iin len() ilevi kullann  ardndan 0'dan balayn ve dizinlerine bavurarak tuple eleri arasnda dng yapn.

Her yinelemeden sonra dizini 1 artrmay unutmayn.

```python
i=0
tuple1=("a","b","c")

while i<len(tuple1):
  print(tuple1[i])
  i=i+1
```
