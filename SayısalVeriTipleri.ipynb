# **Sayısal Veri Tipleri (NUMBERS)**

Pythonda sayısal 3 tip veri tipi vardır
1. int
2. float
3. complex

image.png

Sayısal değişkenler onlara bir değer atandığında oluşturulurlar.

```python
x=1
y=1.3
z=1j
```

Pythonda değişkenlerin tipini sorgulamak için type() metodu kullanılır.

```python
print(type(x))
print(type(y))
print(type(z))
```

## 1. İnt Veri Tipi
Sınırsız uzunlukta,pozitif veya negatif tam sayılardır.

```python
x=1
y=334422343434
z=-123

print(type(x))
print(type(y))
print(type(z))
```

## 2. Float Veri Tipi

Ondalıklı sayılardır.

```python
x = 35e3
y = 12E4
z = -87.7e100

print(type(x))
print(type(y))
print(type(z))
```

## 3. Complex Veri Tipi
Karmaşık sayılardır.
Sanal kısmı j ile yazılır.

```python
x = 3+5j
y = 5j
z = -5j

print(type(x))
print(type(y))
print(type(z))
```

# **Tür Dönüşümü**
Değişkenleri bir türden başka bir türe dönüştürebiliriz.

1. int() : değişkenin tipini int'e çevirir.
2. float() : değişkenin tipini float'a çevirir.
3. complex() : değişkenin tipini complex'e çevirir.


```python
#intten floata dönüşüm
x=1
a=float(x)
print(a)
print(type(a))
```

```python
#intten complex'e dönüşüm
x=1
a=complex(x)
print(a)
print(type(a))
```

```python
#floattan int'e dönüşüm
x=2.3
a=int(x)
print(a)
print(type(a))
```

```python
#floattan complex'e dönüşüm
x=2.3
a=complex(x)
print(a)
print(type(a))
```

####NOT!
Complex sayılar diğer türlere dönüştürülemezler.
